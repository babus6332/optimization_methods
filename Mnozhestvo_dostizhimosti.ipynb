```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
from math import sqrt, cos, sin, atan2
```

```python
def x(t, u_0, tau, a, K, M):
    w = sqrt(2*a)
    if u_0 == 1:
        
        R_1 = sqrt((a*K/w + 1/w)**2 + M**2)
        Theta_1 = atan2(M, a*K/w + 1/w)
        
        L = R_1*(w/a)*cos(w*tau + Theta_1) - 1/a
        P = R_1*sin(w*tau + Theta_1)
        
        R_2 = sqrt((a*L/w - 1/w)**2 + P**2)
        Theta_2 = atan2(P, a*L/w - 1/w) - w*tau
        
        if 0 <= t and t <= tau:
            x_1 = R_1*(w/a)*cos(w*t + Theta_1) - 1/a
            x_2 = R_1*sin(w*t + Theta_1)
            return np.array([x_1, x_2])
        elif tau < t and t <= 1:
            x_1 = R_2*(w/a)*cos(w*t + Theta_2) + 1/a
            x_2 = R_2*sin(w*t + Theta_2)
            return np.array([x_1, x_2])
        
    elif u_0 == -1:
        R_1 = sqrt((a*K/w - 1/w)**2 + M**2)
        Theta_1 = atan2(M, a*K/w - 1/w)
        
        L = R_1*(w/a)*cos(w*tau + Theta_1) + 1/a
        P = R_1*sin(w*tau + Theta_1)
        
        R_2 = sqrt((a*L/w + 1/w)**2 + P**2)
        Theta_2 = atan2(P, a*L/w + 1/w) - w*tau

        if 0 <= t and t <= tau:
            x_1 = R_1*(w/a)*cos(w*t + Theta_1) + 1/a
            x_2 = R_1*sin(w*t + Theta_1)
            return np.array([x_1, x_2])
        elif tau < t and t <= 1:
            x_1 = R_2*(w/a)*cos(w*t + Theta_2) - 1/a
            x_2 = R_2*sin(w*t + Theta_2)
            return np.array([x_1, x_2])
```

```python
a = 3
K = 1
M = 1
tau_array = np.linspace(0, 1, 101)
x_1_lim = []
x_2_lim = []
for u_0 in [-1, 1]:
    for tau in tau_array:
        #t_array = np.linspace(0, 1, 101)
        #x_array = np.array([x(t, u_0, tau, a, K, M) for t in t_array])
        #x_1_lim.append(x_array[-1][0])
        #x_2_lim.append(x_array[-1][1])
        x_lim = x(1, u_0, tau, a, K, M)
        x_1_lim.append(x_lim[0])
        x_2_lim.append(x_lim[1])
plt.figure(figsize = (7, 7))
plt.plot(x_1_lim, x_2_lim)
plt.scatter(3, 1)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.title('Множество достижимости системы (а = {})'.format(a))
plt.plot(np.linspace(-5, 5, 21), [0]*21, '--', c = 'black', alpha = 0.5)
plt.plot([0]*21, np.linspace(-5, 5, 21), '--', c = 'black', alpha = 0.5)
plt.show()
```
